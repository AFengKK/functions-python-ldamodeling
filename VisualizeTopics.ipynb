```python
import pyLDAvis.gensim
import pickle
from urllib.request import urlopen
from gensim import corpora
```

```python
LDA_MODEL_BLOB_URL = "https://ldamodel101.blob.core.windows.net/ldamodel/ldamodel"
TOKEN_DATA_URL = "https://ldamodel101.blob.core.windows.net/ldamodel/token_data"
```

```python
# Load the trained model and the token map from blob
loaded_pickle_object = pickle.load(urlopen(LDA_MODEL_BLOB_URL)) 
token_data = pickle.load(urlopen(TOKEN_DATA_URL))
```

```python
# Build dictionary and corpus using token data
dictionary = corpora.Dictionary(token_data)
corpus = [dictionary.doc2bow(text) for text in token_data]
```

```python
# Call PyLDAvis to prepare the visualization
vis = pyLDAvis.gensim.prepare(loaded_pickle_object, corpus, dictionary)
```

```python
# Show the visualization
pyLDAvis.display(vis)
```
